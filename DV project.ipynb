#cleaning data

```python
# drop unnecessary column
MD=us_accident.loc[us_accident['State'] == 'MD']
MD=MD.drop('Number',axis=1)
print(MD)
MD.isnull().sum()
```

```python
def impute_median(series):
    return series.fillna(series.median())

def impute_mean(series):
    return series.fillna(series.mean())
```

```python
# impute mdeian for continuous data
MD.loc[:,'End_Lat'] = MD.loc[:,'End_Lat'].transform(impute_median)
MD.loc[:,'End_Lng'] = MD.loc[:,'End_Lng'].transform(impute_median)
MD.loc[:,'TMC'] = MD.loc[:,'TMC'].transform(impute_median)
MD.loc[:,'Temperature(F)'] = MD.loc[:,'Temperature(F)'].transform(impute_median)
MD.loc[:,'Wind_Chill(F)'] = MD.loc[:,'Wind_Chill(F)'].transform(impute_median)
MD.loc[:,'Humidity(%)'] = MD.loc[:,'Humidity(%)'].transform(impute_median)
MD.loc[:,'Pressure(in)'] = MD.loc[:,'Pressure(in)'].transform(impute_median)
MD.loc[:,'Visibility(mi)'] = MD.loc[:,'Visibility(mi)'].transform(impute_median)
MD.loc[:,'Wind_Speed(mph)'] = MD.loc[:,'Wind_Speed(mph)'].transform(impute_median)
MD.loc[:,'Precipitation(in)'] = MD.loc[:,'Precipitation(in)'].transform(impute_median)
```

```python
MD.isnull().sum()
```

```python
# impute mode for categorical data
MD = MD.fillna(MD['Weather_Condition'].value_counts().index[0])
MD = MD.fillna(MD['Astronomical_Twilight'].value_counts().index[0])
MD = MD.fillna(MD['Nautical_Twilight'].value_counts().index[0])
MD = MD.fillna(MD['Weather_Timestamp'].value_counts().index[0])
MD = MD.fillna(MD['Civil_Twilight'].value_counts().index[0])
MD = MD.fillna(MD['City'].value_counts().index[0])
MD = MD.fillna(MD['Wind_Direction'].value_counts().index[0])
MD = MD.fillna(MD['Sunrise_Sunset'].value_counts().index[0])
MD = MD.fillna(MD['Zipcode'].value_counts().index[0])
MD = MD.fillna(MD['Timezone'].value_counts().index[0])
MD = MD.fillna(MD['Airport_Code'].value_counts().index[0])
MD = MD.fillna(MD['Description'].value_counts().index[0])
```

```python
# make sure there is no missing value
MD.isnull().sum()
```
